```python
#use this
from PIL import Image
import PIL.ImageOps  
import numpy as np
import pandas as pd
import os
import cv2
import csv
Image.MAX_IMAGE_PIXELS = None
from tqdm import tqdm
import operator
import matplotlib.pyplot as plt
from tifffile import imread
%matplotlib inline
GT ='GT'
Assess = 'Assess'
```

```python
def dir_create(path):
    if not os.path.exists(path):
        os.makedirs(path)
        
def makepseudo(arr,minpercent=5,maxpercent=95):
    minval,maxval = np.percentile(arr.flatten(),[minpercent,maxpercent])
    if(maxval>minval):
        arr = np.maximum(np.minimum(arr,maxval),minval)
        arr = arr-minval
        arrf = np.array(arr,dtype=np.float32)
        arrf = np.array(np.round(255.0*np.minimum(1.0,arrf/(maxval-minval))),dtype=np.uint8)
    else:
        arrf = np.array(arr[:,:],dtype=np.uint8)
    return(arrf)

def make_merge_channels_8bit(r,g):
    b = np.zeros(r.shape)
    bgr_uint8 = (np.dstack((b,g,r)) ) .astype(np.uint8) 
    return bgr_uint8

def red_green_image_maker(r,g):
    red_green_image= make_merge_channels_16bit(makepseudo(r),makepseudo(g))
    return red_green_image


def show(arr):
    im = Image.fromarray(np.array(255*arr/np.max(arr),dtype=np.uint8))
    im.show()

def getcentre(contours):
    M = cv2.moments(contours)
    cX = int(M["m10"] / M["m00"])
    cY = int(M["m01"] / M["m00"])
    return((cX,cY))

def are_same_fibres(cont2, cont1):
    c1X,c1Y = getcentre(cont1)
    #c2X,c2Y = getcentre(cont2)
    dist = cv2.pointPolygonTest(cont2,(c1X,c1Y),True)
    if dist > 0:
        val = True
    else:
        val = False
    return val

def drawcentres(img_msk_blend, arr,contours,labels=[],ptrad=3):
    '''Modified function from mitocyto to write number and centres on contours'''
    rgb = img_msk_blend
    h,w = arr.shape
    uselabs = True 
    
    for i,cnt in enumerate(contours):
        cX,cY = getcentre(cnt)
        cv2.circle(rgb, (cX,cY), ptrad, (255, 0, 0), -1)
        if uselabs and arr.shape[1]>100:
            cv2.putText(rgb, str(i), (min(w-10,max(10,cX - 20)), min(h-10,max(10,cY - 10))),cv2.FONT_HERSHEY_SIMPLEX, 0.5*arr.shape[1]/2656.0, (255, 255, 255), 1)
          
    return(rgb)


def metric_cal(img, annarr,contour,element_size=6): #2 for IMC 6 for QIF
    element = cv2.getStructuringElement(cv2.MORPH_RECT, (2 * element_size + 1, 2 * element_size + 1),
                                        (element_size, element_size))
    
    black = np.zeros(annarr.shape,dtype=np.uint8)
    cv2.drawContours(black,[contour], -1, (255), -1)
    edge_eroded = black - cv2.erode(black,element) # area to check for dystrophin +ve
    edge_dilated = cv2.dilate(black,element) - black  # area to check for VDAC +ve
        
        # calculating pixels in three area 1) dilated area 2) eroded area and 3) fibre area - eroded area
    outer_pixels = edge_dilated == 255
    inner_pixels = edge_eroded == 255
    fibre_pixels = black == 255
    check_area = outer_pixels + inner_pixels
        #show(check_area)
        #show(inner_pixels)

     # greater of memb and mass pixels   
    #memb_pixels= img[:,:,2] - img[:,:,1]
    #membrane_thresh = memb_pixels > 0
    
    #mass_pixels= img[:,:,1] -img[:,:,2]
    #mito_mass_thresh = mass_pixels > 0
    
    
    membrane_thresh = img[:,:,2] > np.quantile(img[:,:,2],0.85)
    mito_mass_thresh = img[:,:,1] > np.quantile(img[:,:,1],0.75)
   
    
   
        
        # calculating red(dystophin) and green(VDAC) pixels in appropriate areas
    membrane_included = np.logical_and(membrane_thresh, inner_pixels)
    mito_mass_missed = np.logical_and(mito_mass_thresh, outer_pixels)
    
        
        # calculating proportions
    proportion_membrane_included = (np.sum(membrane_included)/np.sum(inner_pixels))*100
    proportion_mito_mass_missed = (np.sum(mito_mass_missed)/np.sum(outer_pixels))*100
    
        
    values = [proportion_membrane_included,proportion_mito_mass_missed]
    return values

def IoU_two_contours(ann,cont1, cont2):
    contours = [cont1, cont2]

        # Create image filled with zeros the same size of original image
    blank = np.zeros(ann.shape[0:2])

        # Copy each contour into its own image and fill it with '1'
    image1 = cv2.drawContours(blank.copy(), contours, 0, 255,thickness = cv2.FILLED) # 0 index contour i.e. first one
    image2 = cv2.drawContours(blank.copy(), contours, 1, 255,thickness = cv2.FILLED) # 1 index contour i.e. second one

        # Use the logical AND operation on the two images
        # Since the two images had bitwise and applied to it,
        # there should be a '1' or 'True' where there was intersection
        # and a '0' or 'False' where it didnt intersect
    intersection = np.logical_and(image1, image2)
    union = np.logical_or(image1, image2)
    IoU = np.sum(intersection)/np.sum(union)    
          
    return IoU

def contour_finder (mask):
    
    # Find fibres from mask_GT
    contours,hierarchy = cv2.findContours(mask, cv2.RETR_CCOMP,2)
    # Ensure only looking at holes inside contours...
    contours = [c for i,c in enumerate(contours) if hierarchy[0][i][3] != -1]
    return contours

def contour_finder_instance_mask (instance_mask):
    
    instance_mask[0,:-1] = instance_mask[:-1,-1] = instance_mask[-1,::-1] = instance_mask[-2:0:-1,0] = instance_mask.max()
    unique_colours = np.unique(instance_mask[np.nonzero(instance_mask)])
    contours = []
    for colour in unique_colours:
        unique_colour_contour_mask = instance_mask == colour
        unique_colour_contour_mask= np.array(unique_colour_contour_mask,dtype=np.uint8)
        
        
        thresh, unique_colour_contour_binary  = cv2.threshold(unique_colour_contour_mask, 0, 255, cv2.THRESH_BINARY)
        unique_colour_contour_binary = 255-unique_colour_contour_binary
        conturs = contour_finder (unique_colour_contour_binary)
        for contour in conturs:
            contours.append(contour)
                   
    return contours
    

    
    

def IoU_score (element_size,img, annarr, mask_GT, mask_Assess, is_gt_instance_mask, is_assess_instance_mask):
    
    annarr[0,:-1] = annarr[:-1,-1] = annarr[-1,::-1] = annarr[-2:0:-1,0] = annarr.max()
    
    
 
    
    if is_gt_instance_mask:
        contours_GT= contour_finder_instance_mask(mask_GT)
    else:
        mask_GT= 255-mask_GT
        mask_GT[0,:-1] = mask_GT[:-1,-1] = mask_GT[-1,::-1] = mask_GT[-2:0:-1,0] = mask_GT.max()
        contours_GT= contour_finder (mask_GT)
        
    
    if is_assess_instance_mask:
        contours_Assess= contour_finder_instance_mask(mask_Assess)
    else:
        mask_Assess= 255-mask_Assess
        mask_Assess[0,:-1] = mask_Assess[:-1,-1] = mask_Assess[-1,::-1] = mask_Assess[-2:0:-1,0] = mask_Assess.max()
        contours_Assess= contour_finder(mask_Assess)
        
    #print(len(contours_Assess), len(contours_Assess[0]))
        
        
    
    
   
    results = [] 
    total_GT_contours= len(contours_GT)
    total_Assess_contours= len(contours_Assess)
    for i,contour in enumerate(tqdm((contours_GT), desc='Contours processed')):
        counter = 0
        for j, cont in enumerate(contours_Assess):
            
            
            
            val = are_same_fibres(contours_GT[i],contours_Assess[j])
            if val :
                if counter != 0:
                    results.pop()
                GT_annotation= metric_cal(img, annarr,contours_GT[i],element_size=element_size)
                Assess_annotation= metric_cal(img, annarr,contours_Assess[j],element_size=element_size)
                IoU = IoU_two_contours(annarr,contours_GT[i], contours_Assess[j])
                values = [i,j,total_GT_contours,total_Assess_contours,IoU,GT_annotation[0],GT_annotation[1],Assess_annotation[0],Assess_annotation[1]]
                results.append(values)
                
                break
            
            elif counter == 0:
                place_holder= 0
                fiber_match_not_found= 'not_found'
                GT_annotation= metric_cal(img, annarr,contours_GT[i],element_size=element_size)
                values = [i,fiber_match_not_found,total_GT_contours,total_Assess_contours, place_holder,GT_annotation[0],GT_annotation[1],place_holder,place_holder]
                results.append(values)
                counter += 1
                
            
                
    return results
                
```

```python
inp_img_dir = './sample_data/sample_IF_images/'
inp_GT_msk_dir = './sample_data/sample_IF_groundtruth_masks/'
inp_assess_msk_dir = './sample_data/sample_assess_masks/'
out_dir = './sample_data/sample_outputs/quality_metrics_csvs/'


dir_create(out_dir)
element_size = 6 # set 2 for IMC or 6 for QIF


img_list = [f for f in
            os.listdir(inp_img_dir)
            if os.path.isfile(os.path.join(inp_img_dir, f))]

for infile in img_list:
    infile_img_path = os.path.join(inp_img_dir, infile)
    infile_msk_GT_path = os.path.join(inp_GT_msk_dir, infile.split('.')[0] + '.tiff')
    infile_msk_Assess_path = os.path.join(inp_assess_msk_dir, infile.split('.')[0] + '.tiff')
    #infile_out_path = os.path.join(out_dir, infile.split('.')[0] + '_Asses.csv')
    infile_out_path_IoU = os.path.join(out_dir, infile.split('.')[0] + '.csv')
    
   
    img = cv2.imread(infile_img_path)
    mask_Assess = imread(infile_msk_Assess_path) # change load type to 2 for 16 or 32 bit mask otherwise 0
    #mask_Assess = np.array(mask_Assess_t ,dtype=np.uint8)
    #thresh,mask_Assess = cv2.threshold(mask_Assess, 0, 255, cv2.THRESH_BINARY)
    mask_GT = imread(infile_msk_GT_path) # change load type to 2 for 16 or 32 bit mask otherwise 0
    
    annarr = np.array(mask_GT,dtype=np.uint8)
    thresh,annarr = cv2.threshold(annarr, 0, 255, cv2.THRESH_BINARY)    
    
    results_IoU = IoU_score (element_size,img,annarr, mask_GT, mask_Assess,is_gt_instance_mask=True, is_assess_instance_mask=True)# change instance mask boolean
   
    
    
    header1 = ['Fibre_'+GT,'Fibre_'+Assess,'Total_'+GT+'contours','Total_'+Assess+'_contours', 'IoU_Score_'+Assess+'_'+GT,GT+'_proportion_membrane_included', GT+'_proportion_mito_mass_missed',Assess+'_proportion_membrane_included', Assess+'_proportion_mito_mass_missed' ]
    with open(infile_out_path_IoU, "w", encoding="UTF8",newline='') as f:
        writer = csv.writer(f)
        # write the header
        writer.writerow(header1)
        # write the data
        writer.writerows(results_IoU)
```
